```python
# Import Modules
from ACC_functions import *
from pathlib import Path
import pandas as pd
```

```python
# Import CSV files
dir = Path.cwd() / "logs" 
df = (pd.read_csv(files, index_col=0) for files in dir.glob("*.csv"))
df = pd.concat(df)
working_df = df
```
